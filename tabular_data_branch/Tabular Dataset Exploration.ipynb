### Preprocess matt_metadata

<ol>
    <li>Encode categorical features with numbers</li>
    <li>Normalize matt_metadata</li>
    <li>Fill NaN's with 0.5</li>
</ol>

```python
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        pass
    try:
        import unicodedata
        unicodedata.numeric(s)
        return True
    except (TypeError, ValueError):
        pass
    return False
```

```python
def norm(read_from, write_to):
    w = open(write_to, 'w', encoding='utf-8')
    f = open(read_from, 'r')
    
    list_header = f.readline().rstrip('\r\n').split('\t') # Abbrev. column names
    w.write('\t'.join(list_header) + '\n')
    list_norm_range = f.readline().rstrip('\r\n').split('\t') # Normal range
#     w.write('\t'.join(list_norm_range) + '\n')
    list_norm_range = list_norm_range[1:]
    
    dic={}
    for l in f: # Read all patients
        sp = l.rstrip('\r\n').split('\t')
        dic[sp[0]] = sp
    
    for key in dic: # For each patient
        flist = []
        sex = ['Male', 'Female']
        for i, ft in enumerate(dic[key][1:]):
            if i == 0:flist.append(ft) # Age
            if i == 1: flist.append(sex.index(ft)) # Gender
            if i == 2: # Temperature
                if is_number(ft):flist.append(float(ft) / 37.2) # Divide by avg. human body temperature
                else:flist.append(1)
            if i == 3: # Udis
                if ft in ['No underlying disease']:flist.append(0) 
                else:flist.append(1)
            if i > 3: # All other features -> Min-max normalization
                # Extract min and max
                if '-' in list_norm_range[i]: # If normal range of this feature contains '-'
                    max = float(list_norm_range[i].split(' ')[0].split('-')[1])
                    min = float(list_norm_range[i].split(' ')[0].split('-')[0])
                elif '<' in list_norm_range[i]: # If normal range of this feature contains '<'
                    max = float(list_norm_range[i].split(' ')[0].split('<')[1])
                    min = 0
                
                ft = ft.replace('>', '').replace('<', '')
                
                # Min-max normalization
                if is_number(ft):flist.append((float(ft) - min) / (max - min))
                else:flist.append(0.5)

        w.write(key  + '\t' + '\t'.join([str(s) for s in flist]) + '\n')
```

```python
norm(read_from="matt_metadata_unnorm.csv",
     write_to="matt_metadata_norm.csv")
```

### Split train and val sets

```python
full_df = pd.read_csv("matt_metadata_norm_mortality.csv", sep='\t')
```

```python
full_df
```

```python
deceased_df = full_df.loc[full_df["Mortality"] == 1]
cured_df = full_df.loc[full_df["Mortality"] == 0]
```

```python
train_deceased_idx = list(deceased_df.sample(n=53).index)
train_cured_idx = list(cured_df.sample(n=450).index)
```

```python
val_deceased_idx = list(set(deceased_df.index) - set(train_deceased_idx))
val_cured_idx = list(set(cured_df.index) - set(train_cured_idx))
```

```python
with open("../train_idx.csv", "w") as f:
    for idx in train_cured_idx + train_deceased_idx:
        f.write(f"{idx}\n")
```

```python
with open("../val_idx.csv", "w") as f:
    for idx in val_cured_idx + val_deceased_idx:
        f.write(f"{idx}\n")
```
